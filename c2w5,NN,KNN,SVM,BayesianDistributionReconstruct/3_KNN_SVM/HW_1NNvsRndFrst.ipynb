```python
from sklearn.datasets import load_digits
```

```python
data_dg = load_digits()
```

```python
X_train = data_dg.data[:int(0.75*data_dg.data.shape[0])]
y_train = data_dg.target[:int(0.75*data_dg.data.shape[0])]

X_test = data_dg.data[int(0.75*data_dg.data.shape[0]):]
y_test = data_dg.target[int(0.75*data_dg.data.shape[0]):]

print(data_dg.data.shape)
print(X_train.shape[0] + X_test.shape[0])
print(y_train.shape[0] + y_test.shape[0])
```

```python
import numpy as np

def KNN_1k_predict(Xt,X,y):
    pred = [sorted(zip(((xt-X)**2).sum(axis=1),y))[0][1] for xt in Xt]
    #pred = [0 for i in range(Xt.shape[0])]
    #for i,xt in enumerate(Xt):
#        pred[i] = sorted(zip(((xt-X)**2).sum(axis=1),y))[0][1]
    return pred
```

```python
KNN_1k_predict(X_test, X_train, y_train)
```

```python
answ_1 = sum(KNN_1k_predict(X_test, X_train, y_train) != y_test) / y_test.shape[0]
```

```python
print(answ_1)
```

```python
def write_answer(answ, num_):
    with open("answer_"+str(num_)+".txt", "w") as fout:
        fout.write(str(answ))
```

```python
write_answer(answ_1, 1)
```
